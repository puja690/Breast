## Converting catagorial value into Numerical value

```python
dg = pd.get_dummies(breast["diagnosis"],drop_first=True) ##get_dummies  is a function of panda which is used to convert the catogrical value into numerical value
breast=pd.concat([breast,dg],axis=1)
breast.head()
```
